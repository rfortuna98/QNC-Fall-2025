```python
#we figure out that p=0.3
# we change temperature and see that observable release is
k=7
# on still the same number of possible quanta
n=14
#what is the probability of seing 7 or more quanta in a p=0.3
#P(K≥7∣n=14,p=0.3)
from scipy.stats import binom
n=14
k=7
p_null = 0.3  # null hypothesis: release probability didn't change
# Compute P(K >= 7 | n=14, p=0.3)
# This is the survival function (1 - CDF up to k-1)
p_value = binom.sf(k - 1, n, p_null)  # same as 1 - binom.cdf(k - 1, n, p)

print(f"P(K >= {k} | p = {p_null}) = {p_value:.4f}")
#chance is 0.093, above a stipulated threshold of 0.05, so "no significant"

#cool graph chatgpt made!
import matplotlib.pyplot as plt
import numpy as np

k_values = np.arange(0, n + 1)
pmf = binom.pmf(k_values, n, p_null)

plt.bar(k_values, pmf, color='lightgray', edgecolor='black', label='P(k | p=0.3)')
plt.bar(k_values[k_values >= k], pmf[k_values >= k], color='red', label='P(k ≥ 7)')
plt.axvline(k, color='black', linestyle='--', label='Observed k=7')

plt.xlabel('Number of quanta released (k)')
plt.ylabel('Probability')
plt.title('Binomial Distribution under Null Hypothesis (p=0.3)')
plt.legend()
plt.grid(True)
plt.show()
```
