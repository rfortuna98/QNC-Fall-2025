```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Define the dataset
data = {
    "Age": [3, 4, 5, 6, 7, 8, 9, 11, 12, 14, 15, 16, 17],
    "Wing Length": [1.4, 1.5, 2.2, 2.4, 3.1, 3.2, 3.2, 3.9, 4.1, 4.7, 4.5, 5.2, 5.0]
}

df = pd.DataFrame(data)

# Perform linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(df['Age'], df['Wing Length'])

# Create regression line
df['Predicted'] = intercept + slope * df['Age']

# Plot the data and regression line
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Wing Length', data=df, label='Data')
plt.plot(df['Age'], df['Predicted'], color='red', label='Regression Line')
plt.xlabel("Age")
plt.ylabel("Wing Length")
plt.title("Age vs. Wing Length with Regression Line")
plt.legend()
plt.grid(True)
plt.show()

# Print regression info
print(f"Slope: {slope:.4f}")
print(f"Intercept: {intercept:.4f}")
print(f"R-squared (RÂ²): {r_value**2:.4f}")
print(f"Pearson's r: {r_value:.4f}")
print(f"P-value: {p_value:.7f}")
print(f"Standard Error: {std_err:.4f}")
```
